```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import cmocean.cm as cmo
import seaborn as sns
sns.set_style('darkgrid')
```

## Read in data

```python
df = pd.read_csv('../../data/atmospheric_forcing/forcing.csv',index_col=0)
df
```

## Plot all climatologies

```python
#%% Plots of all climatologies
months = ['J','F','M','A','M','J','J','A','S','O','N','D']
thermal = matplotlib.cm.get_cmap('cmo.thermal')
ice = matplotlib.cm.get_cmap('cmo.ice')
solar = matplotlib.cm.get_cmap('cmo.solar')
algae = matplotlib.cm.get_cmap('cmo.algae')

def plotClim(ax,variable,unit,color='k'):
    ax.scatter(range(12),df[variable],color=color)
    ax.plot(range(12),df[variable],color=color)
    ax.set_title(variable,fontsize=14)
    if unit!='':
        ax.set_ylabel(variable+' ('+unit+')',fontsize=11)
    else:
        ax.set_ylabel(variable,fontsize=11)
    ax.set_xticks(range(12),months)

fig, ((ax1,ax2,ax3),(ax4,ax5,ax6),(ax7,ax8,ax9)) = plt.subplots(3,3,figsize=(15,10))
plotClim(ax1,'Shortwave radiation','W/m$^2$',solar(0.3))
plotClim(ax2,'Longwave radiation','W/m$^2$',thermal(0.7))
plotClim(ax3,'Air temperature','\u00B0'+'C',thermal(0.9))
plotClim(ax4,'Relative humidity','',thermal(0.4))
plotClim(ax5,'Snow albedo','',ice(0.6))
plotClim(ax6,'Snowfall','m/s',ice(0.8))
plotClim(ax7,'Wind speed','m/s',algae(0.6))
plotClim(ax8,'Wind standard deviation','m/s',algae(0.2))
plotClim(ax9,'Ice/wind ratio','',thermal(0.6))
plt.subplots_adjust(hspace=0.25,wspace=0.35)
plt.savefig('../../atmosphere_plots/atmospheric_forcing_clim.png',bbox_inches='tight',dpi=300)
```
